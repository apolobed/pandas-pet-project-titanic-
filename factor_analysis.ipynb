```python
import pandas as pd
import numpy as np

result = pd.read_excel("data.xlsx")

result.head()
```

```python
properties_result = result.iloc[:, 4:].copy()
properties_result.head()
```

```python
corr_matrix_pd = properties_result.corr()
corr_matrix = corr_matrix_pd.values

index_to_delete = []

for i in range(len(corr_matrix)):
    for j in range(len(corr_matrix[i])):
        if i != j:
            if abs(corr_matrix[i][j])>0.7:
                index_to_delete.append([i, j])


excluded_names = ["ЧСС/ПАНО (пульс)", "Общий энергет. фонд(усл.един.)"]
cleaned_results = result.drop(columns=excluded_names)
cleaned_results.head()
```
